```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import LabelEncoder
```

# loading dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values    
```

```python

labelencoder = LabelEncoder()
dataset['Position'] = labelencoder.fit_transform(dataset['Position'])
```

# Splitting the dataset into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
linear = LinearRegression()
linear.fit(X_train, y_train)
```

# Visualizing the Simple Linear Regression results

```python
plt.scatter(X, y, color='red')
plt.plot(X, linear.predict(X), color='blue')
plt.title('Simple Linear Regression')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

# Polynomial Linear Regression

```python
polynomial = PolynomialFeatures(degree=4)
X_poly = polynomial.fit_transform(X)
linear_2 = LinearRegression()
linear_2.fit(X_poly, y)
```

# Visualizing the Polynomial Regression results

```python
plt.scatter(X, y, color='red')
plt.plot(X, linear_2.predict(polynomial.fit_transform(X)), color='green')
plt.title('Polynomial Regression')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```
